```python
# 노이즈 데이터('Unknown') 제거
headlines = [line for line in df.headline.values if line != 'Unknown']
len(headlines)
```

```python
# encoding, 구두점 제거, 소문자로 변환
import re
def preprocessing(s):
    s = s.encode('utf8').decode('ascii', 'ignore')
    s = re.sub('[^A-Za-z0-9 ]', '', s)
    return s.lower()
```

```python
headlines = [preprocessing(line) for line in headlines]
headlines[:5]
```

- Keras 텍스트 전처리

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
```

```python
# 단어 집합 생성
t = Tokenizer()
t.fit_on_texts(headlines)
```

```python
# 단어 집합 크기
vocab_size = len(t.word_index) + 1
vocab_size
```

```python
# sequence 생성
sequences = []
for line in headlines:
    encoded = t.texts_to_sequences([line])[0]
    for i in range(1, len(encoded)):
        sequences.append(encoded[:i+1])
sequences[:10]

# sequences = []
# for line in text.split('\n'):
#     encoded = t.texts_to_sequences([line])[0]
#     for i in range(1, len(encoded)):
#         sequences.append(encoded[:i+1])
# sequences
```

```python
# sequence 중에서 제일 길이가 긴 것
max_len = max(len(seq) for seq in sequences)
max_len
```

```python
sequences = pad_sequences(sequences, maxlen=max_len, padding='pre')
sequences[:3]
```

```python
X = sequences[:,:-1]
y = sequences[:,-1]
Y = to_categorical(y)
X.shape, Y.shape
```

### 모델 정의/설정/학습
- Embediing vector 개수: 10
- LSTM unit 개수: 128

```python
import numpy as np
import tensorflow as tf
seed = 2022
np.random.seed(seed)
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential([
    Embedding(vocab_size, 10, input_length = max_len -1),
    LSTM(128),
    Dense(vocab_size, activation='softmax')
])
model.summary()
```

```python
model.compile('adam', 'categorical_crossentropy', ['accuracy'])
```

```python
%time hist = model.fit(X, Y, epochs=200, verbose=0)
```

```python
hist.history['accuracy'][-1]
```
